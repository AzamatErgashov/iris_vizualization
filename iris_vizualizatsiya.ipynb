```python
cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
```

# Legend qo'shish uchun biz Histogram quruvchisiga ustun nomlari ro'yxati sifatida teglar qo'shishimiz va legand funksiyani chaqirishimiz kerak.

```python
plt.hist([data["sepal-length"],data["sepal-width"],data["petal-length"],data["petal-width"]], label=[cols[0],cols[1],cols[2],cols[3]])
plt.legend()
```

# plt.legend funksiyasi matplotlib kutubxonasi ichidagi bir funksiya bo'lib, ko'rsatmalarni chizishda yordam beradi. Bu funksiya yordamida chizilgan grafikalar ustida ko'rsatmalar yaratish mumkin. Masalan, agar siz bir nechta liniyali grafik chizmoqchi bo'lsangiz, plt.legend funksiyasi yordamida har bir liniya uchun mos ko'rsatma yaratib, ularni ajratib turishingiz mumkin. Bu, grafikni tushunarliroq qilishga yordam beradi.

# Yoki biz Histogram quruvchisiga 4 ta alohida chaqirishimiz va 4 ta bir-biriga o'xshash chizmalarni olishimiz mumkin.

```python
plt.hist(data["sepal-length"])
plt.hist(data["sepal-width"]) 
plt.hist(data["petal-length"])
plt.hist(data["petal-width"])
```

# biz histogramlarni shaffofligini boshqarish uchun alfadan foydalanishimiz mumkin. 

# alfani 1 shaffof qilib beradi ya'ni alfa 1 gacha son yozsanggiz boladi masal 0.5,0.9 
# 1 ni uzi esa shaffolik yoq degan manoni bildiradi

```python
plt.hist(data["sepal-length"])
plt.hist(data["sepal-width"],alpha=1) 
plt.hist(data["petal-length"],alpha=0.6)
plt.hist(data["petal-width"],alpha=0.3)
```

Shuningdek, biz uni yanada o'qilishi uchun 4 ta ustunni alohida pastki qismlarga chizishimiz mumkin. Bu juda o'qilishi mumkin, lekin ehtiyot bo'ling, har bir uchastka avtomatik ravishda o'z o'qlarini ma'lumotlarga o'tkazadi.

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 10))
ax[0, 0].hist(data["sepal-length"])
ax[0, 1].hist(data["sepal-width"])
ax[1, 0].hist(data["petal-length"])
ax[1, 1].hist(data["petal-width"])
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 8))
ax[0, 0].hist(data["sepal-length"],color="lime")
ax[0, 1].plot(data["sepal-width"])
ax[1, 0].plot(data["petal-length"],c="m")
ax[1, 1].boxplot(data["petal-width"])

ax[0, 0].set_title(cols[0])
ax[0, 1].set_title(cols[1])
ax[1, 0].set_title("petal-leght")
ax[1, 1].set_title(cols[3])
plt.show()
```

# Scatter Plots

## Bu maʼlumotlar toʻplami uchun foydaliroq boʻlishi mumkin, chunki ular turlar boʻyicha klasterlarni koʻrsatishi mumkin. Eng asosiy tarqalish diagramma turlarni ajratmaydi.

```python
plt.scatter(
    data["petal-length"], 
    data["petal-width"],)
```

# Turlar bo'yicha rang kodlashni qo'shish bizga har bir tur uchun 2 ta atribut uchun klasterlashni ko'rish imkonini beradi. Bu erda setosa tanho, lekin virginica va versikolor bir-biriga mos keladi.

```python
colors = {'Iris-setosa':'red',"Iris-virginica":'m',"Iris-versicolor":'blue'}
plt.scatter(
    data['sepal-length'],
    data["sepal-length"],
    c=data["class"].map(colors)
)
# plt.legend()
plt.show()
```

# X va y o'qlariga teglar qo'shish foydalidir, ammo biz virginica va versicolor ma'lumotlarining bir-biriga mos kelishini ko'rishimiz mumkin. Agar biz ushbu 2 tur uchun bir-biriga mos kelmaydigan 1 atributni topsak, ularni aniq ajratish uchun ulardan foydalanishimiz mumkin. Ammo, afsuski, barcha to'rtta atribut bir-biriga mos keladi.

```python
plt.scatter(
    data['sepal-length'],
    data['petal-length'], 
    c=data['class'].map(colors))
plt.xlabel(cols[0])
plt.ylabel(cols[2])
```

# plt.title title yozib beradi

```python
plt.scatter(
    data['sepal-length'],
    data['petal-length'], 
    c=data['class'].map(colors))
plt.xlabel(cols[0])
plt.ylabel(cols[2])
plt.title('iris data vizualizatsiya')
```

# Box and Whisker Plots

# Qutidagi chizmalar 25, 50 (o'rtacha) va 75 foizlarni ko'rsatish orqali atribut bo'yicha ma'lumotlarning taqsimlanishini ko'rsatadi. Shunga qaramay, eng oddiy syujetlar unchalik foydali emas, lekin biz teglar va rang kodini qo'shganda, uchastkalar ochiladi.

```python
plt.boxplot([data["sepal-length"],data["sepal-width"],data["petal-length"],data["petal-width"]])
```

<h4>Quticha chizmasining o'rtasidagi quti (box) ma'lumotlar to'plamingizning 25, 50 va 75 percentillarini ko'rsatadi. Chizmaning pastki va yuqori chizi esa ma'lumotlar to'plamingizning minimum va maksimum qiymatlarini ko'rsatadi.

<h3>percentillar ma'lumotlar to'plamingizning qanday tarzda joylashganligi va o'rtacha qiymatlari haqida ma'lumot beradi.

<h2>25, 50, va 75 percentillar ma'lumotlar to'plamingizni o'rtacha qiymatlarga bo'lib bo'lishadi

# Bu turlar boʻyicha saralangan bitta atribut uchun quti grafigini koʻrsatadi. Ushbu atribut uchun biz 3 tur o'rtasida katta o'xshashlikni ko'rishimiz mumkin, shuning uchun u farqlash uchun unchalik foydali emas. Ushbu atribut uchun 5,5 yoki 6,0 bo'lgan iris 3 turdagi har qanday bo'lishi mumkin.

```python
data.boxplot(column=["sepal-length"], by='class')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(9, 8))
A = [data["sepal-length"][data["class"] == 'Iris-setosa'], data["sepal-length"][data["class"] == 'Iris-virginica'], data["sepal-length"][data["class"] == 'Iris-versicolor']]
B = [data["sepal-width"][data["class"] == 'Iris-setosa'], data["sepal-width"][data["class"] == 'Iris-virginica'], data["sepal-width"][data["class"] == 'Iris-versicolor']]
C = [data["petal-length"][data["class"] == 'Iris-setosa'], data["petal-length"][data["class"] == 'Iris-virginica'], data["petal-length"][data["class"] == 'Iris-versicolor']]
D = [data["petal-width"][data["class"] == 'Iris-setosa'], data["petal-width"][data["class"] == 'Iris-virginica'], data["petal-width"][data["class"] == 'Iris-versicolor']]

ax[0, 0].boxplot(A, widths = 0.7)
ax[0, 0].set_title(cols[0])
ax[0, 1].boxplot(B, widths = 0.7)
ax[0, 1].set_title(cols[1])
ax[1, 0].boxplot(C, widths = 0.7)
ax[1, 0].set_title(cols[2])
ax[1, 1].boxplot(D, widths = 0.7)
ax[1, 1].set_title(cols[3])
```
